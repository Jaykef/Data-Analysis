```python
import sqlite3
import pandas as pd
from sqlalchemy import create_engine
import matplotlib.pyplot as plt
```

```python
input_file_1 = 'shift-data.xlsx'
input_file_2 = 'third-shift-data.xlsx'
output_file = 'output_data.xlsx'
```

```python
df_first_shift = pd.read_excel(input_file_1, sheet_name='first')
df_second_shift = pd.read_excel(input_file_1, sheet_name='second')
df_third_shift = pd.read_excel(input_file_2)

df_all = pd.concat([df_first_shift,df_second_shift,df_third_shift])

```

```python
engine = create_engine('sqlite://', echo=False)
df_all.to_sql('employees', engine, if_exists='replace', index=False)

results = engine.execute("SELECT * FROM employees")

final = pd.DataFrame(results, columns=df_all.columns)
final.to_excel(output_file, index=False)
final
```
